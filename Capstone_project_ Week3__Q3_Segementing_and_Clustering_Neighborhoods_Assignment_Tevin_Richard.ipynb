#### 3.6 Get the top 50 venues that are within a 500 meter radius 

```python
# define the radius
radius = 500 

# define the limit of the results
LIMIT = 50 

venues = []

for latitude, longitude, post, borough, neighbourhood in zip(complete_dataset['Latitude'], complete_dataset['Longitude'], complete_dataset['Postal Code'], complete_dataset['Borough'], complete_dataset['Neighbourhood']):
    url = "https://api.foursquare.com/v2/venues/explore?client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}".format(
        CLIENT_ID,
        CLIENT_SECRET,
        VERSION,
        latitude,
        longitude,
        radius, 
        LIMIT)
    
    results = requests.get(url).json()["response"]['groups'][0]['items']
    
    for venue in results:
        venues.append((
            post, 
            borough,
            neighbourhood,
            latitude, 
            longitude, 
            venue['venue']['name'], 
            venue['venue']['location']['lat'], 
            venue['venue']['location']['lng'],  
            venue['venue']['categories'][0]['name']))
```

#### 3.10 Use one hot encoding to analyze and evaluate each area

```python
# Initialise the one hot encoding
venue_onehot = pd.get_dummies(venue_df[["Venue Category"]], prefix="", prefix_sep="")

# Add the Borough and neighbourhood name to the data frame
venue_onehot["Borough"] = venue_df["Borough"]
venue_onehot["Neighbourhood"] = venue_df["Neighbourhood"]

# Move the Borugh and Neighbourhood columns from the end of the data frame to the start for uniformity
aranged_columns = list(venue_onehot.columns[-2:]) + list(venue_onehot.columns[:-2])
venue_onehot = venue_onehot[aranged_columns]

venue_onehot
```

#### 3.11 Group the venue data by neighourhood

```python
venue_onehot_grouped = venue_onehot.groupby("Neighbourhood").mean().reset_index()
venue_onehot_grouped
```

#### 3.12 Get the tope 10 venues for each neighbourhood

```python
# create a function that will check for top 10

def most_common_venues(row, number_of_top_venues_required):
    row_cat = row.iloc[1:]
    row_cat_sorted = row_cat.sort_values(ascending=False)
    
    return row_cat_sorted.index.values[0:number_of_top_venues_required]
```

```python
import numpy as np
```

```python
# create the dataframe with the top 10 common venues in each neighbourhood

# Make a list to indicate the ranking (st for 1st, nd for 2nd and rd for 3rd)
indicators = ["st", "nd", "rd"]
Number_of_top_venues_required = 10 

# start creating columns for the data frame

columns = ["Neighbourhood"]

for ind in np.arange(Number_of_top_venues_required):
    try:
        columns.append("{}{} Most Common Venue ". format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'. format(ind+1))
        
sorted_neighbourhood_venues = pd.DataFrame(columns=columns)
sorted_neighbourhood_venues["Neighbourhood"] = venue_onehot_grouped["Neighbourhood"]

for ind in np.arange(venue_onehot_grouped.shape[0]):
    sorted_neighbourhood_venues.iloc[ind, 1:] = most_common_venues(venue_onehot_grouped.iloc[ind, :], Number_of_top_venues_required)
    

sorted_neighbourhood_venues.head()
        
```

#### 3.13 Cluster the neighbourhoods using K-Means



```python
# import required libraries
from sklearn.cluster import KMeans as KMS
```

```python
# Define the number of clusters to generate
kms_clusters = 5
```

```python
# Create a new dataframe using the grouped dataframe and drop the Neighbourhood colum so that new data frame only has venue catergories
group_clustering_toronto = venue_onehot_grouped.drop("Neighbourhood",1)
```

```python
# Create the Kmeans clustering model and fit the data

kms_ = KMS(n_clusters = kms_clusters, random_state=0).fit(group_clustering_toronto)
```

```python
# get the clustering labels for the data
kms_.labels_
```

#### 3.14 Add the cluster labels to the data frame


```python
sorted_neighbourhood_venues. insert(0, "Clusering Labels", kms_.labels_)
```

#### 3.17 Plot the clusters on the map of Toronto

```python
import matplotlib.cm as cm
import matplotlib.colors as colors

# create map
toronto_clusters = folium.Map(location=[latitude, longitude], zoom_start=10)

# set color scheme for the clusters
x = np.arange(kms_clusters)
ys = [i+x+(i*x)**2 for i in range(kms_clusters)]
colors_array = cm.rainbow(np.linspace(0, 1, len(ys)))
rainbow = [colors.rgb2hex(i) for i in colors_array]

# add markers to the map
markers_colors = []
for lat, lon, post, bor, poi, cluster in zip(combined_data_final['Borough Latitude'], combined_data_final["Borough Longitude"], combined_data_final['Postal Code'], combined_data_final['Borough'], combined_data_final['Neighbourhood'], combined_data_final['Clusering Labels']):
    label = folium.Popup('{} ({}): {} - Cluster {}'.format(bor, post, poi, cluster), parse_html=True)
    folium.CircleMarker(
        [lat, lon],
        radius=5,
        popup=label,
        color=rainbow[cluster-1],
        fill=True,
        fill_color=rainbow[cluster-1],
        fill_opacity=0.6).add_to(toronto_clusters)
       
toronto_clusters
```
