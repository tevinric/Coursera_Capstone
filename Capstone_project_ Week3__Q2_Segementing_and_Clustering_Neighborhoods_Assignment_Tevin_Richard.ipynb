# Question 1 

#### 1.1 Import the required libraries

```python
import pandas as pd
import requests 
```

#### 1.2. Extract the required data from the wikipedia page

```python
url ="https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"

#save data to variable using the .get method

wikipedia_data = requests.get(url)   
```

#### 1.3 Load the page html text data into a pandas dataframe

```python
#using the .text method to pull the text data

table_data = pd.read_html(wikipedia_data.text)

print("The loaded data is a ", type(table_data), "with ", len(table_data), "elements")
```

```python
# Take a look at the data
table_data
```

#### 1.4 Extract the data required (First element in the list containing our three columns) 

```python
# replce the list with our required data since lists are mutable
table_data = table_data[0]
table_data
```

#### 1.5 Drop all the Boroughs that contain Not Assigned elements

```python
df = table_data[table_data["Borough"] != "Not assigned"]
df
```

#### 1.6 Group all the elements that have the same postal codes as a single entry

```python
df = df.groupby(['Postal Code']).head()
df
```

#### 1.7.Fix the indices

```python
# reset the dataframe index
df = df.reset_index()
df
```

```python
# replace the old index with the updated indices
df = df.drop(["index"], "columns")
df
```

#### 1.8 Get the shape of the data set

```python
print("The data frame has ", df.shape[0], "rows and ", df.shape[1], "columns")
```
